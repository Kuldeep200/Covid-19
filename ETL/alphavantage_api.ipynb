```python
# Dependencies
import pandas as pd
import json
import requests
from config import api_key
import time
```

```python
# This gets the monthly inflation expectation data
url = 'https://www.alphavantage.co/query?function=INFLATION_EXPECTATION&apikey=api_key'
r = requests.get(url)
inflation_data = r.json()

print(inflation_data)
```

```python
inflation_df = pd.json_normalize(inflation_data, record_path=['data'])
inflation_df
```

```python
# export to CSV file
inflation_df.to_csv('Resources/raw data files/monthly_inflation.csv')
```
