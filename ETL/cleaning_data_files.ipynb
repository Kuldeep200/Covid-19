## cleaning the Covid-19 data

```python
file="Resources/raw data files/covid cases daily.xlsx"
df_covid_cases=pd.read_excel(file)
df_covid_cases.head()
```

```python
# covert to datetime
df_covid_cases['Date_reported'] = pd.to_datetime(df_covid_cases['Date_reported'], errors='coerce')
# show the types 
df_covid_cases.info()
```

```python
df_covid_cases['Month']= df_covid_cases['Date_reported'].dt.month
df_covid_cases.head()
```

```python
df_covid_cases['Year']= df_covid_cases['Date_reported'].dt.year
df_covid_cases.head()
```

```python
cols=["Month","Year"]
df_covid_cases['Date']=df_covid_cases[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis='columns')
df_covid_cases.head()
```

```python
df_covid_cases.drop(df_covid_cases.columns[[0, 1, 2, 4, 6, 7, 8]], axis = 1, inplace = True)
df_covid_cases.head()
```

```python
df_covid_cases.shape
```

```python
new_df = df_covid_cases.groupby('Date').sum().reset_index()
new_df
```

```python
# add in a day to the date column
```

```python
# Save new dateframe as CSV file
# export to CSV file
new_df.to_csv('Resources/cleaned_covid_data.csv')
```

## cleaning the housing data

```python
file="Resources/cleaned_housing_data.csv"
df_houses=pd.read_csv(file)
df_houses.head()
```

```python
# covert to datetime
df_houses['Month of Period End'] = pd.to_datetime(df_houses['Month of Period End'], errors='coerce')
# show the types 
df_houses.info()
```

```python
df_houses['Month']= df_houses['Month of Period End'].dt.month
df_houses.head()
```

```python
df_houses['Year']= df_houses['Month of Period End'].dt.year
df_houses.head()
```

```python
cols=["Month","Year"]
df_houses['Date']=df_houses[cols].apply(lambda x: '-'.join(x.values.astype(str)), axis='columns')
df_houses.head()
```

```python
df_houses.drop(df_houses.columns[[0, 1, 4, 5]], axis = 1, inplace = True)
df_houses.head()
```

```python
# convert the median sale price to an integer
# remove the dollar signs
df_houses["Median Sale Price"] = df_houses["Median Sale Price"].replace({'\$':''}, regex = True)
# remove the "K"
df_houses["Median Sale Price"] = df_houses["Median Sale Price"].replace({'K':''}, regex = True)
df_houses.dtypes
```

```python
# multiply the median sale price by 1000
df_houses[['Median Sale Price']] = df_houses[['Median Sale Price']].apply(pd.to_numeric)
df_houses[['Median Sale Price']] = df_houses[['Median Sale Price']] * 1000
df_houses.head()
```

```python
# Save new dateframe as CSV file
# export to CSV file
df_houses.to_csv('Resources/cleaned_housing_data.csv')
```
